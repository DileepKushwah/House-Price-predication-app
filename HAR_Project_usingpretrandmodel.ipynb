```python
# Step 1: Import necessary libraries
from transformers import AutoImageProcessor, AutoModelForImageClassification
from PIL import Image
import torch

# Step 2: Load the pre-trained model and processor
processor = AutoImageProcessor.from_pretrained("DunnBC22/vit-base-patch16-224-in21k_Human_Activity_Recognition")
model = AutoModelForImageClassification.from_pretrained("DunnBC22/vit-base-patch16-224-in21k_Human_Activity_Recognition")

# Step 3: Load and preprocess the input image
# Replace with the actual path of your image
image_path = '/content/drive/MyDrive/test/Image_1.jpg'
image = Image.open(image_path)

# Preprocess the image using the processor
inputs = processor(images=image, return_tensors="pt")

# Step 4: Make a prediction
with torch.no_grad():
    outputs = model(**inputs)

# Step 5: Get the predicted label
logits = outputs.logits  # Get the raw prediction scores (logits)
predicted_class_idx = torch.argmax(logits, dim=-1)  # Get the class with the highest score
predicted_class = model.config.id2label[predicted_class_idx.item()]  # Convert index to label

# Step 6: Print the result
print(f"Predicted class: {predicted_class}")
```
